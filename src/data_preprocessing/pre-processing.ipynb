Data quality detection & Data cleaning

```python
# This function is used to check the percentage of missing values in each column of dataframe

def check_null_percentage(df, field_list):
    if df is None:
        print('df does not exist')
        return
    
    if not field_list or len(field_list) == 0:
        print('field list does not exist or is empty')
        return
    
    percent_dic = {}
    for field in field_list:
        null_count =df[field].isnull().sum()
        total_count = len(df[field].index)
    
        if total_count == 0:
            percent =  '0%'
        else:
            percent = str(null_count / total_count * 100) + '%'
            
        percent_dic[field] = percent
        
    return percent_dic
    
    
```

```python
# Null check for each dataframe

# character data frame
print(check_null_percentage(character_df, character_headers))

# movie data frame
print(check_null_percentage(movie_df, movie_headers))

# movie_sentence_frame
print(check_null_percentage(movie_sentence_df, movie_sentence_headers))

# movie_conversation_frame
print(check_null_percentage(movie_dialog_df, movie_dialog_headers))
```
